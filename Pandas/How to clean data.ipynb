```python
import pandas as pd
inventory = pd.read_csv(r"C:\Users\user\Downloads\products-100.csv")
inventory.head()
```

```python
print("to print missing values in each column")
print(inventory.isnull().sum())
```

```python
#handling the missing values by filling it with the median of values
inventory['Price']=inventory['Price'].fillna(inventory['Price'].median())
```

```python
print("to print missing values in each column")
print(inventory.isnull().sum())
# Every column will be filled
```

```python
#to check duplicates
print(inventory.duplicated().sum())
```

```python
inventory=inventory.drop_duplicates()
#to drop duplicates if any
```

```python
inventory.info()
```

```python
#type conversion
inventory['Stock']=inventory['Stock'].astype(float)
```

```python
inventory.info()
```

```python
#ensuring price  is never negative
inventory=inventory[inventory['Price']>=0]
```
